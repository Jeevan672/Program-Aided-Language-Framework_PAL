```python
from langchain_experimental.pal_chain import PALChain
from langchain import OpenAI
```

```python
from kaggle_secrets import UserSecretsClient
user_secrets = UserSecretsClient()
secret_value_0 = user_secrets.get_secret("OPEN_API_KEY")
```

```python
llm = OpenAI(openai_api_key=secret_value_0,temperature=0)
```

```python
pal_chain = PALChain.from_math_prompt(llm, verbose=True)
```

```python
question1=': Roger has 5 tennis balls. He buys 2 more cans oftennis balls. Each can has 3 tennis balls. How many tennis balls does he have now?'
```

```python
pal_chain.invoke(question1)
```

```python
question2=' The bakers at the Beverly Hills Bakery baked 200loaves of bread on Monday morning. They sold 93 loavesin the morning and 39 loaves in the afternoon. A grocerystore returned 6 unsold loaves. How many loaves of breaddid they have left?'
```

```python
pal_chain.invoke(question2)
```

```python
pal_chain = PALChain.from_colored_object_prompt(llm, verbose=True)
```
